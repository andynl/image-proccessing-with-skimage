```python
from skimage import data, color
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_image(image, title='Image', cmap_type='gray'):
  plt.imshow(image, cmap=cmap_type)
  plt.title(title)
  plt.axis('off')
  plt.show()
```

```python
gambar_kopi = data.coffee()
gambar_koin = data.coins()
```

```python
show_image(gambar_kopi)
```

```python
show_image(gambar_koin)
```

```python
rocket = data.rocket()
show_image(rocket)
```

```python
rocket_gray = color.rgb2gray(rocket)
show_image(rocket_gray)
```

```python
pic = rocket

fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15, 15))

for c, ax in zip(range(3), ax):
  
  split_img = np.zeros(pic.shape, dtype="uint8")

  split_img[:, :, c] = pic[:, :, c]

  ax.imshow(split_img)
```

```python
red_channel = rocket[:, :, 0]

plt.hist(red_channel.ravel(), bins=256)

plt.title('Red Histogram')

plt.show()
```

```python
green_channel = rocket[:, :, 1]

plt.hist(green_channel.ravel(), bins=256)

plt.title('Green Histogram')

plt.show()
```

```python
blue_channel = rocket[:, :, 2]

plt.hist(blue_channel.ravel(), bins=256)

plt.title('Blue Histogram')

plt.show()
```

```python
koin_1 = data.coins()
show_image(koin_1)
```

```python
from skimage.filters import threshold_otsu

koin_grayscale = color.rgb2gray(koin_1)

thresh = threshold_otsu(koin_grayscale)

thresh
```

```python
binary = koin_grayscale > thresh

show_image(binary)
```

```python
from skimage.filters import sobel
```

```python
koin_sobel = sobel(koin_grayscale)

show_image(koin_sobel)
```

```python
from skimage.filters import gaussian
```

```python
gaussian_kopi = gaussian(gambar_kopi, multichannel=True)

show_image(gaussian_kopi)
```

```python
from skimage import exposure

gambar_bagus = exposure.equalize_adapthist(gambar_kopi, clip_limit=0.3)

show_image(gambar_bagus)
```
